```python
import pandas as pd
import matplotlib.pyplot as plt

# load data
df = pd.read_csv('data.csv')  # replace 'data.csv' with the path to your CSV file

# convert the time to a more readable format
df['time'] = pd.to_datetime(df['time'], unit='ns')

# calculate elapsed time in minutes
df['time'] = (df['time'] - df['time'].iloc[0]).dt.total_seconds() / 60.0

# plot data
plt.figure(figsize=(10,6))
plt.plot(df['time'], df['charge_level'])
plt.xlabel('Elapsed Time (minutes)')
plt.ylabel('Charge Level')
plt.title('Charge Level Over Time')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
